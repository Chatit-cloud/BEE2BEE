## 2. Start a Node
A node is your entrypoint into the network.

```python
import asyncio
from bee2bee import P2PNode

async def start():
    node = P2PNode(port=0)
    await node.start()
    print(f"Node ID: {node.peer_id}")
    print(f"Address: {node.addr}")
    return node

# Note: In Jupyter, you might need to run this in a background task if you want to keep the cell free
loop = asyncio.get_event_loop()
node = loop.create_task(start())
```

## 3. Discovering Peers
Once connected to a bootstrap, you can see all available AI providers.

```python
providers = node.result().list_providers()
print(f"Found {len(providers)} providers on the mesh.")
```
